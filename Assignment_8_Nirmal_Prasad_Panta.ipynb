### c) Now compute each of the six quantities above and compare to your answers in a) and b).

*ANS:*
***

1. $$\text{Var}[y_{i,j} \mid \theta_j, \sigma^2] = \sigma^2
$$
(variance of data normally distributed with mean $\theta_j$ and variance $\sigma^2$)
---
2. $$
\text{Var}(\bar{y}_{\cdot, j} \mid \theta_j, \sigma^2) = \text{Var} \left[ \frac{1}{n_j} \sum_{i=1}^{n_j} y_{i,j} \mid \theta_j, \sigma^2 \right]
$$

$$
= \frac{1}{n_j^2} \, \text{Var} \left[ \sum_{i=1}^{n_j} y_{i,j} \mid \theta_j, \sigma^2 \right]
$$

$$
= \frac{1}{n_j^2} \sum_{i=1}^{n_j} \text{Var} \left[ y_{i,j} \mid \theta_j, \sigma^2 \right]
$$
(Since the $y_{i,j}$ are independent given $\theta_j$)

$$
= \frac{1}{n_j^2} (n_j \sigma^2)
$$

$$
= \frac{\sigma^2}{n_j}
$$

---
3. $$
\text{Cov}[y_{i_1 j}, y_{i_2 j} | \theta_j, \sigma^2] = \mathbb{E}[y_{i_1 j} y_{i_2 j} | \theta_j, \sigma^2] - \mathbb{E}[y_{i_1 j} | \theta_j, \sigma^2] \mathbb{E}[y_{i_2 j} | \theta_j, \sigma^2]
$$

Since $y_{i_1 j}$ and $y_{i_2 j}$ are independent given $\theta_j$,
$$
\mathbb{E}[y_{i_1 j} y_{i_2 j} | \theta_j, \sigma^2] = \mathbb{E}[y_{i_1 j} | \theta_j, \sigma^2]\mathbb{E}[y_{i_2 j} | \theta_j, \sigma^2]
$$

Therefore,
$$
\text{Cov}[y_{i_1 j}, y_{i_2 j} | \theta_j, \sigma^2] = \mathbb{E}[y_{i_1 j} | \theta_j, \sigma^2]\mathbb{E}[y_{i_2 j} | \theta_j, \sigma^2] - \mathbb{E}[y_{i_1 j} | \theta_j, \sigma^2]\mathbb{E}[y_{i_2 j} | \theta_j, \sigma^2] = 0
$$


---

4. $$
\text{Var}[y_{i,j} | \mu, \tau^2]
$$


$$
\text{Var}[y_{i,j} | \mu, \tau^2]
= \mathbb{E}[\text{Var}(y_{i,j} | \theta_j, \sigma^2)] + \text{Var}(\mathbb{E}[y_{i,j} | \theta_j, \sigma^2])
$$
(Since, law of total variance:
$
\text{Var}[Y] = \mathbb{E}[\text{Var}(Y | X)] + \text{Var}(\mathbb{E}[Y | X])
$
$$
= \mathbb{E}[\sigma^2] + \operatorname{Var}(\theta_j)
$$

$$
= \sigma^2 + \tau^2
$$

---

5. $$\text{Var}[\bar{y}_{\cdot, j} | \mu, \tau^2]
$$

$$
= \text{Var} \left[ \frac{1}{n} \sum_{i} y_{i, j} \mid \mu, \tau^2 \right]
$$

Applying the law of total variance:
$$
= \mathbb{E} \left[ \text{Var} \left( \frac{1}{n} \sum_{i} y_{i, j} \mid \theta_j, \sigma^2 \right) \mid \mu, \tau^2 \right]
  + \text{Var} \left( \mathbb{E} \left[ \frac{1}{n} \sum_{i} y_{i, j} \mid \theta_j, \sigma^2 \right] \mid \mu, \tau^2 \right)
$$

For $y_{i, j} | \theta_j, \sigma^2$ independent and identically distributed, from 2
$$
\text{Var} \left( \frac{1}{n} \sum_{i} y_{i, j} \mid \theta_j, \sigma^2 \right) = \frac{\sigma^2}{n}
$$

$$
\mathbb{E} \left[ \frac{1}{n} \sum_{i} y_{i, j} \mid \theta_j, \sigma^2 \right] = \theta_j
$$
so,
$$
\text{Var} \left( \mathbb{E} \left[ \frac{1}{n} \sum_{i} y_{i, j} \mid \theta_j, \sigma^2 \right] \mid \mu, \tau^2 \right) = \text{Var}( \theta_j \mid \mu, \tau^2 ) = \tau^2
$$

Hence,
$$
\text{Var}[\bar{y}_{\cdot, j} | \mu, \tau^2] = \frac{\sigma^2}{n} + \tau^2
$$

---

6.

$$
\text{Cov}[y_{i_1 j}, y_{i_2 j} \mid \mu, \tau^2]
= \mathbb{E}\left[\text{Cov}\left(y_{i_1 j}, y_{i_2 j} \mid \theta_j, \sigma^2\right)\right]
+ \text{Cov}\left(\mathbb{E}[y_{i_1 j} \mid \theta_j, \sigma^2], \mathbb{E}[y_{i_2 j} \mid \theta_j, \sigma^2]\right)
$$

From 3, the first covariance term is $0$. So,

$$
\mathbb{E}[0] + \text{Cov}[\theta_j, \theta_j]
$$

$$=\text{Cov}[\theta_j, \theta_j]$$
$$= \text{Var}[\theta_j] $$
$$= \tau^2$$.

Hence,
$$
\text{Cov}[y_{i_1 j}, y_{i_2 j} \mid \mu, \tau^2] = \tau^2

### **8.3 Hierarchical modeling:** The files `school1.dat` through `school8.dat` give weekly hours spent on homework for students sampled from eight different schools. Obtain posterior distributions for the true means for the eight different schools using a hierarchical normal model with the following prior parameters:

### $\mu_0 = 7,\ \gamma_0^2 = 5,\ \tau_0^2 = 10,\ \eta_0 = 2,\ \sigma_0^2 = 15,\ \nu_0 = 2.$

### a) Run a Gibbs sampling algorithm to approximate the posterior distribution of $\{\theta, \sigma^2, \mu, \tau^2\}$. Assess the convergence of the Markov chain, and find the effective sample size for $\{\sigma^2, \mu, \tau^2\}$. Run the chain long enough so that the effective sample sizes are all above 1,000.

*ANS:*


```r
install.packages("coda")
install.packages("MCMCpack")
```

```r
y_1 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school1.dat")
y_2 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school2.dat")
y_3 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school3.dat")
y_4 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school4.dat")
y_5 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school5.dat")
y_6 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school6.dat")
y_7 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school7.dat")
y_8 = scan("https://www2.stat.duke.edu/~pdh10/FCBS/Exercises/school8.dat")
```

```r
# data summaries
ybar = c(mean(y_1), mean(y_2), mean(y_3), mean(y_4), mean(y_5), mean(y_6), mean(y_7), mean(y_8))
ni = c(length(y_1), length(y_2), length(y_3), length(y_4), length(y_5), length(y_6), length(y_7), length(y_8))
si2 = c(var(y_1), var(y_2), var(y_3), var(y_4), var(y_5), var(y_6), var(y_7), var(y_8))

## Setting the priors
mu0 = 7; # prior mean of mu
gamma02 = 5; # the prior variance of mu
#
nu0 = 2 # nu0/2 is the shape of prior for sigma^2
sigma02 = 15 # nu0*sigma^2/2 involved in the rate of the prior for sigma^2
#
eta0 = 2 # eta0/2 is the shape of prior for tau^2
tau02 = 10 # eta0*tau0^2/2 involved in the rate of the prior for tau^2

# Initialize the parameters
theta=c(1, 1, 1, 1, 1, 1, 1, 1)
sigma2=1
tau2=1
mu=1

###############################
ntot=sum(ni) # total sample size
emm=length(theta) # number of groups
#
capn=50000 # the no of MCMC samples
#
storetheta=matrix(0, nrow=capn, ncol=8);
storesigma2=rep(0, capn);
storetau2=rep(0, capn);
storemu=rep(0, capn);

# the begining of the MCMC iterations
for(it in 1:capn){

## Sampling of theta
fcd_var_theta= 1/(ni/sigma2+1/tau2)
fcd_mean_theta= fcd_var_theta*(ni*ybar/sigma2+mu/tau2)
theta=rnorm(emm, fcd_mean_theta, sqrt(fcd_var_theta));
## Sampling of sigma2
fcd_rate_sig2=(sum((ni-1)*si2+ ni*(ybar-theta)^2)
+nu0*sigma02)/2
fcd_shape_sig2= (ntot+nu0)/2
sigma2=1/rgamma(1, fcd_shape_sig2, rate=fcd_rate_sig2)
## Sampling tau2
fcd_rate_tau2=(sum((theta-mu)^2)+eta0*tau02)/2
fcd_shape_tau2= (emm+eta0)/2
tau2=1/rgamma(1, fcd_shape_tau2, rate=fcd_rate_tau2)
## Sampling of mu
fcd_var_mu= 1/(emm/tau2+1/gamma02)
fcd_mean_mu= fcd_var_mu*(emm* mean(theta)/tau2+mu0/gamma02)
mu= rnorm(1, fcd_mean_mu, sqrt(fcd_var_mu))
##
storetheta[it, ]=theta
storesigma2[it]=sigma2
storetau2[it]=tau2
storemu[it]=mu
}




```

```r
library(coda)

# Combine everything into one matrix
mcmc_samples <- data.frame(theta1=storetheta[, 1], theta2=storetheta[, 2],
                           theta3=storetheta[, 3], theta4=storetheta[, 4],
                           theta5=storetheta[, 5], theta6=storetheta[, 6],
                           theta7=storetheta[, 7], theta8=storetheta[, 8],
                           mu=storemu, sigma2=storesigma2, tau2=storetau2)

# Convert to mcmc object
mcmc_obj <- as.mcmc(mcmc_samples)
# Plot trace plot for convergence assessment
traceplot(mcmc_obj)
```

```r
# Calculates Effective Sample Size
effectiveSize(mcmc_obj)
```

Hence, we can see from the trace plot that $  \theta, \sigma^2, \mu, \text{and } \tau^2\ $ have converged successfully. For 50,000 iterations, the effective sample size for all parameters are well beyond 1000.
